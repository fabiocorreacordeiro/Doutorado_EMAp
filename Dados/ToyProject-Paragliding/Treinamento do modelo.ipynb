```python
# Carregando o dataset
a_file = open("ToyProject-Dataset.pkl", "rb")
dataset = pickle.load(a_file)
a_file.close()
```

```python
# Separando as chaves do dicionário entre treino (70%), validação (20%) e teste (10%).

dataset_keys = list(dataset.keys())
random.shuffle(dataset_keys)

n = len(dataset_keys)
dataset_train_key = dataset_keys[0:int(n*0.7)]
dataset_val_key = dataset_keys[int(n*0.7):int(n*0.9)]
dataset_test_key = dataset_keys[int(n*0.9):]
```

```python
# Rever a importância de cada atributo após o treinamento do modelo
k = ['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)',
     'PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO, HORARIA (mB)',
     'RADIACAO GLOBAL (KJ/m²)',
     'TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)',
     'TEMPERATURA DO PONTO DE ORVALHO (°C)',
     'UMIDADE RELATIVA DO AR, HORARIA (%)',
     'VENTO_X, RAJADA MAXIMA (m/s)',
     'VENTO_Y, RAJADA MAXIMA (m/s)', 
     'VENTO_X, VELOCIDADE HORARIA (m/s)', 
     'VENTO_Y, VELOCIDADE HORARIA (m/s)'] 
     
j = ['Day sin', 
     'Day cos',
     'Year sin',
     'Year cos']

df_k = pd.DataFrame(map(dataset[dataset_train_key[0]].get, k), index=k)
df_j = pd.DataFrame(map(dataset[dataset_train_key[0]].get, j), index=j)
```
