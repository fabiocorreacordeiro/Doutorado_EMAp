```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

### Carregando dados e limpando os dados

```python
# Carregando os dados dos boletins técnicos do INPE
path = '../Boletins_Tecnicos/'

arquivos = ["Boletins_tecnicos_2008.json", "Boletins_tecnicos_2009.json", "Boletins_tecnicos_2010.json", "Boletins_tecnicos_2011.json",
            "Boletins_tecnicos_2012.json", "Boletins_tecnicos_2013.json", "Boletins_tecnicos_2014.json", "Boletins_tecnicos_2015.json",
            "Boletins_tecnicos_2016.json", "Boletins_tecnicos_2017.json", "Boletins_tecnicos_2018.json", "Boletins_tecnicos_2019.json",
            "Boletins_tecnicos_2020.json"]

boletins = pd.DataFrame()

for arquivo in arquivos:
  
  df = pd.read_json(path + arquivo,
                    orient='index',
                    convert_axes = False)
  
  boletins = pd.concat([boletins, df])

# Ajustando o formato da data
boletins.index = pd.to_datetime(boletins.index, format= '%d/%m/%Y-00Z')
boletins
```

```python
# Carregando os dados das estações meteorológicas

path = '../Estacoes_Meteorologicas/Rio de Janeiro/'

arquivos = ["INMET_SE_RJ_A652_FORTE DE COPACABANA_01-01-2008_A_31-12-2008.CSV",
            "INMET_SE_RJ_A652_FORTE DE COPACABANA_01-01-2009_A_31-12-2009.CSV",
            "INMET_SE_RJ_A652_FORTE DE COPACABANA_01-01-2010_A_31-12-2010.CSV",
            "INMET_SE_RJ_A652_FORTE DE COPACABANA_01-01-2011_A_31-12-2011.CSV",
            "INMET_SE_RJ_A652_FORTE DE COPACABANA_01-01-2012_A_31-12-2012.CSV",
            "INMET_SE_RJ_A652_FORTE DE COPACABANA_01-01-2013_A_31-12-2013.CSV",
            "INMET_SE_RJ_A652_FORTE DE COPACABANA_01-01-2014_A_31-12-2014.CSV",
            "INMET_SE_RJ_A652_FORTE DE COPACABANA_01-01-2015_A_31-12-2015.CSV",
            "INMET_SE_RJ_A652_FORTE DE COPACABANA_01-01-2016_A_31-12-2016.CSV",
            "INMET_SE_RJ_A652_FORTE DE COPACABANA_01-01-2017_A_31-12-2017.CSV",
            "INMET_SE_RJ_A652_FORTE DE COPACABANA_01-01-2018_A_31-12-2018.CSV",
            "INMET_SE_RJ_A652_RIO DE JANEIRO - FORTE DE COPACABANA_01-01-2019_A_31-12-2019.CSV",
            "INMET_SE_RJ_A652_RIO DE JANEIRO - FORTE DE COPACABANA_01-01-2020_A_31-12-2020.CSV"]

estacao = pd.DataFrame()

for arquivo in arquivos:
  
  df = pd.read_csv(path+arquivo, 
                   delimiter=";",
                   header=8,
                   encoding= "iso-8859-1")

  estacao = pd.concat([estacao, df])

estacao = estacao.reset_index()
```

```python
# Ajustando as colunas de Datas
estacao['Data'] = estacao['DATA (YYYY-MM-DD)'].astype(str) + estacao['Data'].astype(str)
estacao['Data'] = estacao['Data'].apply(lambda x: x.replace('nan','').replace('-', '/'))

# Ajustando as colunas de Horas
estacao['HORA (UTC)'] = estacao['HORA (UTC)'].apply(lambda x: str(x).replace(':00','00 UTC') )
estacao['Hora UTC'] = estacao['HORA (UTC)'].astype(str) + estacao['Hora UTC'].astype(str)
estacao['Hora UTC'] = estacao['Hora UTC'].apply(lambda x: x.replace('nan',''))

# Criando coluna UTC - 3 (horário no Rio de Janeiro)
estacao["UTC-3"] = estacao["Hora UTC"].apply(lambda x: int(x[0:4]) - 300)

# Alterando de strig para datetime
estacao['Data'] = pd.to_datetime(estacao['Data'])

# Criando a coluna "Data UTC-3"
estacao['Data UTC-3'] = estacao['Data']

# Atrasando em um dia as linhas com UTC-3 negativos
estacao.loc[(estacao['UTC-3'] < 0, 'Data UTC-3')] = estacao.loc[(estacao['UTC-3'] < 0, 'Data UTC-3')] - timedelta(days=1)

# Somando 2400 (24h) para os valore "UTC-3" negativos
estacao.loc[(estacao['UTC-3'] < 0, 'UTC-3')] = estacao.loc[(estacao['UTC-3'] < 0, 'UTC-3')] + 2400

# Ajustando as colunas de Radiação Global
estacao['RADIACAO GLOBAL (KJ/m²)'] = (estacao['RADIACAO GLOBAL (KJ/m²)'].astype(str) + 
                                      estacao['RADIACAO GLOBAL (Kj/m²)'].astype(str))
estacao['RADIACAO GLOBAL (KJ/m²)'] = estacao['RADIACAO GLOBAL (KJ/m²)'].apply(lambda x: x.replace('nan',''))

#  Função para colocar um zero antes de número com vírgula. Ex: ',1' => '0,1'.
def change_comma(str): 
    if len(str) > 0:        
        if str[0] == ',':
            return '0' + str
        else:
            return str

estacao['RADIACAO GLOBAL (KJ/m²)'] = estacao['RADIACAO GLOBAL (KJ/m²)'].apply(lambda x: change_comma(x))

# Excluindo colunas redundantes
estacao = estacao.drop(['index', 
                        'DATA (YYYY-MM-DD)', 
                        'HORA (UTC)', 
                        'RADIACAO GLOBAL (Kj/m²)', 
                        'Unnamed: 19'], axis=1)
```

```python
# Ajustando os dados das colunas para 'float'
columns = ['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)',
           'PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO, HORARIA (mB)',
           'PRESSÃO ATMOSFERICA MAX.NA HORA ANT. (AUT) (mB)',
           'PRESSÃO ATMOSFERICA MIN. NA HORA ANT. (AUT) (mB)',
           'RADIACAO GLOBAL (KJ/m²)',
           'TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)',
           'TEMPERATURA DO PONTO DE ORVALHO (°C)',
           'TEMPERATURA MÁXIMA NA HORA ANT. (AUT) (°C)',
           'TEMPERATURA MÍNIMA NA HORA ANT. (AUT) (°C)',
           'TEMPERATURA ORVALHO MAX. NA HORA ANT. (AUT) (°C)',
           'TEMPERATURA ORVALHO MIN. NA HORA ANT. (AUT) (°C)', 
           'VENTO, RAJADA MAXIMA (m/s)',
           'VENTO, VELOCIDADE HORARIA (m/s)']

for column in columns:
    estacao[column] = (estacao[column]
                       .str.replace(',', '.', regex=False)
                       .astype(float))
```

```python
# Ajustando o valor da Radiação Global para o período da noite
estacao['RADIACAO GLOBAL (KJ/m²)'] [((estacao['UTC-3'] < 600) | 
                                    (estacao['UTC-3'] > 2000)) & 
                                    (estacao['RADIACAO GLOBAL (KJ/m²)'] < 0)] = 0
```

```python
# Excluindo valores menores do que zero (-9999)
columns = ['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)',
       'PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO, HORARIA (mB)',
       'RADIACAO GLOBAL (KJ/m²)',
       'TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)',
       'TEMPERATURA DO PONTO DE ORVALHO (°C)',
       'UMIDADE RELATIVA DO AR, HORARIA (%)',
       'VENTO, DIREÇÃO HORARIA (gr) (° (gr))', 
       'VENTO, RAJADA MAXIMA (m/s)',
       'VENTO, VELOCIDADE HORARIA (m/s)']

for col in columns:
    estacao = estacao[estacao[col] >= 0]
```

```python
# Carregando dados de vôo - XC Brasil

path = '../XCBrasil/'

arquivo = '8857_Pedra_Bonita'

XC_Brasil = pd.read_csv(path+arquivo, 
                        delimiter=",")
```

```python
# Limpando coluna date
XC_Brasil = XC_Brasil[XC_Brasil['date'] != '00/00/   0']
XC_Brasil['date'] = pd.to_datetime(XC_Brasil['date'])
# Ajustando coluna duration
XC_Brasil['duration'] = XC_Brasil['duration'].apply(lambda x: datetime.strptime(x, '%H:%M'))
# Excluíndo colunas
XC_Brasil = XC_Brasil.drop(['Unnamed: 0'], axis=1)
```

### Unindo os datasets

```python
n = 1000
XC_Brasil['date'][n]
```

```python
dic = XC_Brasil[n:n+1].to_dict('index')
```

```python
est = estacao[estacao['Data UTC-3'] == XC_Brasil['date'][n]]
est = est[(est['UTC-3'] >= 600) & (est['UTC-3'] <= 2000)]
```

```python
columns = ['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)',
       'PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO, HORARIA (mB)',
       'RADIACAO GLOBAL (KJ/m²)',
       'TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)',
       'TEMPERATURA DO PONTO DE ORVALHO (°C)',
       'UMIDADE RELATIVA DO AR, HORARIA (%)',
       'VENTO, DIREÇÃO HORARIA (gr) (° (gr))', 'VENTO, RAJADA MAXIMA (m/s)',
       'VENTO, VELOCIDADE HORARIA (m/s)']

for col in columns:
    dic[n][col] = est[est['Data UTC-3'] == XC_Brasil['date'][n]][col].to_numpy()
```

```python
dic[n]['Nível 250 hPA'] = boletins.loc[XC_Brasil['date'][n]]['Nível 250 hPA']
dic[n]['Nível 500 hPA'] = boletins.loc[XC_Brasil['date'][n]]['Nível 500 hPA']
dic[n]['Nível 850 hPA'] = boletins.loc[XC_Brasil['date'][n]]['Nível 850 hPA']
dic[n]['Superfície'] = boletins.loc[XC_Brasil['date'][n]]['Superfície']
dic[n]['Previsão'] = boletins.loc[XC_Brasil['date'][n]]['Previsão']
```
